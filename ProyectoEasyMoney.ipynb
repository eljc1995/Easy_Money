### una vez visualizados los diferentes df procedemos a unirlos en unos solo
###para esto usaremos un merge

```python
#primero creamos el df0 y hacemos un merge con los df commercial_activity y sociodemographic,
# una vez juntados estos dos, podemos proceder a unir el siguiente df.


df0 = pd.merge(commercial_activity_df,sociodemographic_df,how="right", on=["pk_cid","pk_partition"])
```
